```python
import gensim 
from gensim.models import Word2Vec
Word2Vec
```

```python
import logging
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
#  Keeping the input as a Python built-in list is convenient, 
# but can use up a lot of RAM when the input is large.
# Gensim only requires that the input must provide sentences sequentially, when iterated over. 
# Each sentence a list of words (utf8 strings): 

results = []
with open('DL_machine_learning_sentences_may_2018.txt') as inputfile:
    for line in inputfile:
        #results.append(line.strip().split('\n'))
        results.append(line.lower().split())
results
```

```python
    print (len(results))
```

```python
# build vocabulary and train model
model = gensim.models.Word2Vec(results,size=150,window=10,min_count=2,workers=10)
model.train(results, total_examples=len(results), epochs=10)
```

```python
print (' lunghezza vocab' , '\n' , len(model.wv.vocab),'vocabolario' , list(model.wv.vocab) )
```

```python
model.wv.most_similar("softmax")
```

```python
# text8 corpus
from gensim.models.word2vec import Word2Vec
import gensim.downloader as api

corpus = api.load('text8')  # download the corpus and return it opened as an iterable
model = Word2Vec(corpus)  # train a model from the corpus
model.wv.most_similar("car")
```
