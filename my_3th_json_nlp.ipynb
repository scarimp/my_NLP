FORMATO JSON
-------------
Il formato *json* rappresenta bene una *struttura dati gerarchica*. Quando la si vuole visualizzare in tabella ci sono delle parti che si ripetono. Ricordarisi una tabella per i comuni italiani ( si ripete sempre la regione e la provincia per i comuni appartenenti alla stessa provinci) oppure in un report una serie di costi che appartengono ad una stessa famiglia , tipo costi per viaggi, oppure incassi per interventi tecnici(sviluppare meglio)

```python
import pandas as pd 
import numpy as np 
import json

print ('Version Panda=',pd.__version__)
print ('Version Numpy=',np.__version__)
print ('Version json=',json.__version__)
```

```python
dati = '_sigfox_allan_2018_mini.json'
```

```python
#lettura dati json
json_data = pd.read_json(dati, orient ='columns')
json_data
```

```python
df = pd.DataFrame.from_dict(json_data)
df.head()
```

```python
#serializzazione dati json
ser_json = json_data.to_json(orient="columns")
ser_json
```

```python
#parsing json string
parsed_json = json.loads(ser_json)
parsed_json
```

```python
nomi_tabella = parsed_json['results']['0']['series'][0]['columns']
nomi_tabella
```

```python
valori_tabella =parsed_json['results']['0']['series'][0]['values']
valori_tabella[:7]
```

```python
#index=['d', 'b', 'a'], columns=['two', 'three']
dftabella=pd.DataFrame(valori_tabella,columns=nomi_tabella)
print(dftabella[:7])
```
